Entre outros atributos, a classe `Senador` possui o atributo `materias`, contendo uma lista de todas as matérias cujo autor é o senador referido. Por exemplo, podemos inspecionar a última matéria escrita pelo Senador Renan Calheiros:

Agora, podemos analisar estatísticas sobre as matérias propostas em cada Legislatura. Limitemos o escopo só para as últimas 5 legislaturas.

```python
LEGPATH  = os.path.join(BASEPATH, 'lista', 'legislatura')
legini, legfim = 38, 56
legs = [sn.Legislatura(leg, baselocal=LEGPATH, fetch=True) for leg in range(legini,legfim)]
# Não se faz recursão de busca das informações de cada classe, portanto temos que obtê-las manualmente
for leg in legs:
    for sen in leg.senadores:
        sen.get_info()
print("Adquiridos os dados dos senadores das Legislaturas {}ª à {}ª".format(legini, legfim))
```

Abaixo, avaliamos o número de matérias apresentadas em cada Legislatura, e também a média de matéria/senador.

```python
leg_mats = []
for leg in legs:
    mats = 0
    for sen in leg.senadores:
        for mat in sen.materias:
            if mat.ano >= leg.anoini and mat.ano <= leg.anofim:
                mats += 1
    leg_mats.append(mats)
leg_mats_med = [ leg_mats[i]/len(legs[i].senadores) if len(legs[i].senadores) > 0 else 0 for i in range(len(legs)) ]
```

```python
s_mats = pd.Series(leg_mats, index=[leg.anoini for leg in legs], name='Número de Matérias')
s_mats_med = pd.Series(leg_mats_med, index=s_mats.index, name='Número de Matérias por Senador')

blue = sns.color_palette()[0]
ax1 = s_mats.plot(figsize=(16,4), style='o-')
ax1.set_ylabel(s_mats.name)
ax1.set_xlabel('Ano')
for tl in ax1.get_yticklabels():
    tl.set_color(blue)

ax2 = ax1.twinx()
red = sns.color_palette()[2]
ax2.plot(s_mats_med.index, leg_mats_med, 'o-', color=red)
ax2.set_ylabel('Número de Matérias por Senador')
for tl in ax2.get_yticklabels():
    tl.set_color(red)
ax2.grid(False)
```
