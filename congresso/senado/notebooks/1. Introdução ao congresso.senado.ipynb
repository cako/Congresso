Usando o pacote congresso.senado
--------------------------------

Este pequeno tutorial mostra algumas capacidades que o pacote `congresso.senado` oferece.
Veremos mais adiante como utilizar as classes `Senador` e `Legislatura` para construir alguns gráficos interessantes sobre as composições do Senado através dos anos.

```python
import congresso.senado as sn

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Detalhes de plot, pode ser removido!
%matplotlib inline
import matplotlib as mpl
sns.set_palette("deep", desat=.6)
family, weight = 'Futura Std', 'semibold'
f1, f2, f3, f4 = 22, 18, 14, 10
mpl.rc('font', family=family, weight=weight, size=f2)
mpl.rc('lines', linewidth=2)
mpl.rc('figure', figsize=(16, 9))
mpl.rc('axes', labelsize=f2, titlesize=f1)
mpl.rc('xtick', labelsize=f3)
mpl.rc('ytick', labelsize=f3)
mpl.rc('legend', fontsize=f3)
```

Começando de forma simples, vamos definir um Senador pelo seu código 5322. Faremos também uso da *flag* "*fetch*" para uma busca de seus dados no momento da inicialização. Também são mostrados os atributos públicos da classe, que inclui coisas como o sexo, a data de nascimento (`romario.dn`) etc.

```python
romario = sn.Senador(5322, fetch=True)
list(filter(lambda x: not x.startswith('_'), dir(romario))) # Atributos públicos
```

Se quisermos, por exemplo investigar quantos mandatos Romário teve, podemos simplesmente executar:

```python
len(romario.mandatos)
```

De forma similar, podemos obter os dados sobre Legislaturas. Por exemplo, vejamos a legislatura passada e seus Senadores.

```python
leg54 = sn.Legislatura(54, fetch=True)
list(filter(lambda x: not x.startswith('_'), dir(leg54))) # Atributos públicos
```

```python
[sen.nome for sen in leg54.senadores[:5]]
```

Os dados acima foram todos obtidos através da API pública RESTful que pode ser encontrada [aqui](http://legis.senado.gov.br/dadosabertos/docs/rest.html). Contudo, para não sobrecarregar os servidores e não estar sujeito a velocidade de redes,  é conveniente salvar os dados. Isto é feito usando a *flag* "*write*", e especificando um local para guardar os arquivos. O padrão é o diretório local. Não cobriremos esta parte, mas vamos supor que temos os dados já previamente baixados. Para que estes sejam usados, devemos definir o local onde os arquivos são guardados.

```python
BASEPATH = '~/python/Congresso/congresso/senado/data/senador/'
LEGPATH  = os.path.join(BASEPATH, 'lista', 'legislatura')
```

Podemos agora utilizar os dados locais para compilar uma lista de **todas** as legislaturas até hoje!

```python
legs = [sn.Legislatura(leg, baselocal=LEGPATH, fetch=True) for leg in range(1,56)]
```

E podemos começar a responder algumas perguntas básicas: quantas mulheres temos hoje no Senado? Quantos homens?

```python
nmulheres = sum([1 for s in legs[-1].senadores if s.F ])
nhomens   = sum([1 for s in legs[-1].senadores if s.M ])

print("Número de mulheres na Legislatura atual ({}ª): {}".format(legs[-1].leg, nmulheres))
print("Número de homens na Legislatura atual ({}ª): {}".format(legs[-1].leg, nhomens))
```

Ou, graficamente


```python
df = pd.DataFrame(data=[nhomens, nmulheres], index=['Homens', 'Mulheres'])
df.plot(kind='barh', legend=False, figsize=(16, 4)).set_xlabel("Número de senadores(as)")
```

Um pouco mais complicado, mas não muito, é a distribuição dos sexos através das legislaturas:

```python
n_f = np.array([ sum([1 for s in leg.senadores if s.F ]) for leg in legs ])
n_m = np.array([ sum([1 for s in leg.senadores if s.M ]) for leg in legs ])
```

Para vizualizar gráficamente podemos de novo utilizar a biblioteca `pandas`

```python
df_sexo = pd.DataFrame(data=np.array([n_f, n_m]).T, index=[str(leg.anoini)[2:] for leg in legs],
                       columns=['Mulheres', 'Homens'])
ax = df_sexo.plot(kind='bar',figsize=(18, 4), stacked=True)
for label in ax.xaxis.get_ticklabels():
    label.set_rotation('horizontal')
ax.set_xlabel("Ano inicial da Legislatura (Décadas, começando em 1826)")
ax.set_ylabel("Número de senadores(as)")
```

Para finalizar os exemplos, vamos ver a distribuição dos estados em cada legislatura. Abaixo vamos criar uma uma lista que cada elemento é um dicionário. Esse dicionário contem pares UF - no. de senadores.

```python
leg_UFs = []
for leg in legs:
    UFs = {}
    for sen in leg.senadores:
        for mand in sen.mandatos:
            if mand.legini == leg.leg:
                UFs[mand.UF] = UFs.get(mand.UF, 0) + 1
                break
    leg_UFs.append(UFs)
```

```python
sns.set_palette("deep")
df_ufs = pd.DataFrame(data=leg_UFs, index=[leg.leg for leg in legs])
ax = df_ufs.plot(kind='bar', figsize=(18, 9), stacked=True)
for label in ax.xaxis.get_ticklabels():
    label.set_rotation('horizontal')
ax.set_xlabel("Legislatura")
ax.set_ylabel("Número de senadores(as)")
ax.legend(loc=2, prop={'size':10}, ncol=int(np.ceil(np.sqrt(len(ax.get_legend_handles_labels()[0])))/1.5))
```
